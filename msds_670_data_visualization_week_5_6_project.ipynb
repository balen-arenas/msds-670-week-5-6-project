```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#
```

```python
df = pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
df.head()
```

```python
df['stroke'].value_counts()
```

```python
x = df['age']

fig, ax = plt.subplots()
ax.hist(x, bins=16, linewidth=1, edgecolor='white')
plt.show()
```

```python
df.info()
```

```python
df1  = df.drop(['id', 'gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'], axis=1)
df1.head()
```

```python
ax = sns.heatmap(df1.corr(), annot=True)
```
