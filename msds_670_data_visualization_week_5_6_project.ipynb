```python
mapping = {'never smoked': 0, 'smokes': 1, 'formerly smoked': 2, 'Unknown': 3}
df['smoked'] = df['smoking_status'].map(mapping)
df
```

```python
res_val = {'Urban': 0, 'Rural': 1}
df['residence'] = df['Residence_type'].map(res_val)
df
```

```python
gender_val = {'Male': 2, 'Female': 1, 'Other': 0}
df['gender_value'] = df['gender'].map(gender_val)
df
```

```python
married_val = {'Yes': 1, 'No': 0}
df['married'] = df['ever_married'].map(married_val)
df
```

```python
work_val = {'Private': 0, 'Self-employed': 1, 'Govt_job': 2, 'children': 3, 'Never_worked': 4}
df['work'] = df['work_type'].map(work_val)
df
```

```python
df.info()
```

```python
dfv = df.drop(['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'], axis=1)
dfv.head()
```

```python
dfv = dfv.drop(['id'], axis=1)
dfv.head()
dfv.info()
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
plt.rcParams['font.size'] = 8
plt.rcParams['axes.grid'] = True
plt.style.use('fivethirtyeight')
ax = sns.heatmap(dfv.corr(), annot=True)
```
